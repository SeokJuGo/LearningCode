## 문제

 > 무한히 큰 배열에 다음과 같이 분수들이 적혀있다.

```python
 Image('images/bunsu2.png')
```

```python
n = 4
line = 0
line2 = 0
while(n>line):
    line+=1
    line2+=line
    
    if(line2%2==0): # 짝수라인
        print("짝수")
    else:           # 홀수라인
        print("홀수")
```

이와 같이 나열된 분수들을 1/1 → 1/2 → 2/1 → 3/1 → 2/2 → … 과 같은 지그재그 순서로 차례대로 1번, 2번, 3번, 4번, 5번, … 분수라고 하자.

X가 주어졌을 때, X번째 분수를 구하는 프로그램을 작성하시오.

## 입력

> 첫째 줄에 X(1 ≤ X ≤ 10,000,000)가 주어진다.



## 출력

> 첫째 줄에 분수를 출력한다.

## 예제 입력

> 3 -> 2/1

## 풀이 코드

* 이 문제는 랜덤한 n을 받았을 때 벌집이 몇 겹에 있는지 찾는 문제이다. 벌집이 1을 제외한 6의 배수 단위로 중첩이 되어 += 대입연산자를 사용하여 계속하여 더하여 최소개수를 탐색하는 식으로 문제를 풀었다.

```python
n = int(input())
num = 1  # 1제외하여 6의배수를 더하므로 1부터 시작
count = 1
while n > num :
    num += 6 * count  # 6의 배수씩 계속해서 더하였다.
    count += 1
print(count)
```
